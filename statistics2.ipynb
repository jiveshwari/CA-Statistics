```python
import pandas as pd
import plotly.express as px
from matplotlib import pyplot as plt
```

```python
path = "data/dagstuhl-15512-argquality-corpus-annotated.csv"
file = open(path,encoding='latin-1')

# create df
data = pd.read_csv(file,sep='\t')
    
```

```python
data
```

```python
ann1_low = 0
ann1_avg = 0
ann1_high = 0
ann1_nan = 0

ann2_low = 0
ann2_avg = 0
ann2_high = 0
ann2_nan = 0

ann3_low = 0
ann3_avg = 0
ann3_high = 0
ann3_nan = 0
```

```python
for row in range(0,len(data), 3):
    ann1 = data.iloc[row]
    ann2 = data.iloc[row+1]
    ann3 = data.iloc[row+2]
    if ann1['reasonableness'] == '1 (Low)':
        ann1_low += 1
    if ann1['reasonableness'] == '2 (Average)':
        ann1_avg += 1
    if ann1['reasonableness'] == '3 (High)':
        ann1_high += 1
    if ann1['reasonableness'] != ann1['reasonableness']:
        ann1_nan += 1
        
    if ann2['reasonableness'] == '1 (Low)':
        ann2_low += 1
    if ann2['reasonableness'] == '2 (Average)':
        ann2_avg += 1
    if ann2['reasonableness'] == '3 (High)':
        ann2_high += 1
    if ann2['reasonableness'] != ann2['reasonableness']:
        ann2_nan += 1
        
    if ann3['reasonableness'] == '1 (Low)':
        ann3_low += 1
    if ann3['reasonableness'] == '2 (Average)':
        ann3_avg += 1
    if ann3['reasonableness'] == '3 (High)':
        ann3_high += 1
    if ann3['reasonableness'] != ann3['reasonableness']:
        ann3_nan += 1
```

```python
print("ann1_low ",ann1_low)
print("ann1_avg ",ann1_avg)
print("ann1_high ",ann1_high)
print("ann1_nan ",ann1_nan)

print("ann2_low ",ann2_low)
print("ann2_avg ",ann2_avg)
print("ann2_high ",ann2_high)
print("ann2_nan ",ann2_nan)

print("ann3_low ",ann3_low)
print("ann3_avg ",ann3_avg)
print("ann3_high ",ann3_high)
print("ann3_nan ",ann3_nan)
```

```python
ann1 = (ann1_low,ann1_high,ann1_avg,ann1_nan)
labels =["Low","High","Avg","Nan"]
ann2 = (ann2_low,ann2_high,ann2_avg,ann2_nan)
ann3 = (ann3_low,ann3_high,ann3_avg,ann3_nan)



fig, ax = plt.subplots()
ax.pie(ann1, labels=labels, autopct='%.1f%%')
ax.set_title('Annotator 1')
plt.tight_layout()
fig, ax = plt.subplots()
ax.pie(ann2, labels=labels, autopct='%.1f%%')
ax.set_title('Annotator 2')
plt.tight_layout()
fig, ax = plt.subplots()
ax.pie(ann3, labels=labels, autopct='%.1f%%')
ax.set_title('Annotator 3')
plt.tight_layout()
```
